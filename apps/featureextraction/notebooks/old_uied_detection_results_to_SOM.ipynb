```python
import os
import json
import os

# Tu directorio raíz
raiz = 'D:/01.Tesis/00. platforms/screenrpa-resources/24 Roma Research/v3/'

# Recorrer todos los directorios de la raíz
for dirpath, dirnames, filenames in os.walk(raiz):
    # print(f'Directorio: {dirpath}')
    for dirname in dirnames:
        # Eliminar el directorio con el nombre "decision-tree"
        if dirname == "decision-tree":
            dir_to_remove = os.path.join(dirpath, dirname)
            # Eliminar los archivos dentro del directorio
            for root, dirs, files in os.walk(dir_to_remove, topdown=False):
                for name in files:
                    os.remove(os.path.join(root, name))
                for name in dirs:
                    os.rmdir(os.path.join(root, name))
            # Eliminar el directorio
            os.rmdir(dir_to_remove)
                     
            # Remove the directory from the list
            dirnames.remove(dirname)
    
    # Check if directory exists, if not create it
    if not os.path.exists(dirpath+"_results"):
        os.makedirs(dirpath+"_results")
        if not os.path.exists(os.path.join(dirpath+"_results", "components_json")):
            os.makedirs(os.path.join(dirpath+"_results", "components_json"))
    
    for filename in filenames:
        print(f'Archivo: {filename}')
        if filename.endswith('.json') and not "flattened_dataset" in filename:        
            with open(os.path.join(dirpath, filename), 'r', encoding='utf-8') as f:
                json_f = json.loads(f.read())
                for compo in json_f['compos']:
                    # Add the new key
                    compo["points"] = [[compo["row_min"], compo["column_min"]], [compo["row_max"], compo["column_max"]]]
                    compo["centroid"] = [(compo["row_max"]+compo["row_min"])/2, (compo["column_max"]+compo["column_min"])/2]
                    # Remove other keys
                    compo.pop("column_min", None)
                    compo.pop("row_min", None)
                    compo.pop("column_max", None)
                    compo.pop("row_max", None)
                    compo.pop("width", None)
                    compo.pop("height", None)
                
                
                # Overwrite the file with the new data
                with open(os.path.join(dirpath+"_results", "components_json", filename), 'w', encoding='utf-8') as f:
                    f.write(json.dumps(json_f, indent=4))
        elif "flattened_dataset" in filename:
            # Move the file to os.path.join(dirpath+"_results")
            os.rename(os.path.join(dirpath, filename), os.path.join(dirpath+"_results", filename))
            
```
